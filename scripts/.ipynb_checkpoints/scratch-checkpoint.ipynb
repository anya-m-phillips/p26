```python
n=16
```

```python
init_displacement = init_displacements[n]
pos0 = init_displacement[:3] * u.kpc
vel0 = init_displacement[3:] * (u.km/u.s)
w0 = gd.PhaseSpacePosition(pos0, vel0)
mwp = gp.BovyMWPotential2014(units=galactic)
H = gp.Hamiltonian(mwp)

int_time=10000
dt=1
orbit=H.integrate_orbit(w0, dt=dt, n_steps=int_time/dt)
x_orbit, y_orbit, z_orbit = orbit.pos.x, orbit.pos.y, orbit.pos.z
R_orbit = np.sqrt(x_orbit**2+ y_orbit**2)
```

```python
path = paths[n]
apo = apocenters[n]
```

```python
tdis, tplot_p, tplot_a = paf.find_dissolution_plotting_times(path)
```

```python
# trying to just make this plot one time. 

# clean and prepare data:


i=2555 #int(tdis/2)

all_particles, binaries, singles, coords, single_coords, binary_coords = paf.load_coords(path, i,
                                                                                         return_all_particles=True,
                                                                                         return_single_particles=True,
                                                                                        init_displacement=init_displacements[n])
sx, sy, sz = singles.pos.T * u.pc
sphi1, sphi2, sr, svr = single_coords['phi1'], single_coords['phi2'], single_coords['r'], single_coords['vr']
svx,svy,svz = singles.vel.T * (u.pc/u.Myr)

bx, by, bz = binaries.pos.T * u.pc
bphi1, bphi2, br, bvr = binary_coords['phi1'], binary_coords['phi2'], binary_coords['r'], binary_coords['vr']
bvx,bvy,bvz = binaries.vel.T * (u.pc/u.Myr)

all_x = np.concatenate([sx,bx])
all_y = np.concatenate([sy,by])
all_z = np.concatenate([sz,bz])
# all_vx = np.concatenate([svx,bvx])
# all_vy = np.concatenate([svy,bvy])
# all_vz = np.concatenate([svz,bvz])
all_phi1 = np.concatenate([sphi1,bphi1])
all_phi2 = np.concatenate([sphi2,bphi2])
all_r = np.concatenate([sr,br])
all_vr = np.concatenate([svr,bvr])


### make cuts! 
in_MW = all_r < 1.5 * apo
all_x, all_y, all_z, all_phi1, all_phi2, all_r, all_vr = all_x[in_MW], all_y[in_MW], all_z[in_MW], all_phi1[in_MW], all_phi2[in_MW], all_r[in_MW], all_vr[in_MW]

sigphi1 = paf.calc_dispersion(all_phi1)
phi1mean = np.mean(all_phi1)
clip_phi1 = np.abs(all_phi1 - phi1mean)<=3*sigphi1

sigphi2 = paf.calc_dispersion(all_phi2)
phi2mean = np.mean(all_phi2)
clip_phi2 = np.abs(all_phi2 - phi1mean)<=3*sigphi2

sigr = paf.calc_dispersion(all_r)
rmean = np.mean(all_r)
clip_r = np.abs(all_r - rmean)<=3*sigr

final_clip = clip_phi1 & clip_phi2 & clip_r

all_x, all_y, all_z, all_phi1, all_phi2, all_r, all_vr = all_x[final_clip], all_y[final_clip], all_z[final_clip], all_phi1[final_clip], all_phi2[final_clip], all_r[final_clip], all_vr[final_clip]

### get out of core frame
core = petar.Core(interrupt_mode='bse', external_mode='galpy')
core.loadtxt(path+"data.core")
corepos = core.pos[i] * u.pc
core_x, core_y, core_z = corepos
x_galcen = all_x + core_x
y_galcen = all_y + core_y
z_galcen = all_z + core_z
all_R = np.sqrt(x_galcen**2 + y_galcen**2)

_, straight_vr = paf.straighten_stream_polynomial(all_phi1, all_vr, degree=5, show_plot=False)

##### get plot ready
fig = plt.figure(figsize=(37,15))
gs = GridSpec(2,4, width_ratios=[.5,.5,.5,1], height_ratios=[1,1], hspace=.2, wspace=.3)

ax1 = fig.add_subplot(gs[0,0])
ax1.set_xlabel(r"$x$", fontsize=35)
ax1.set_ylabel(r"$y$", fontsize=35)
ax1.scatter(all_x.to(u.kpc), all_y.to(u.kpc), s=.01, c='k', rasterized=True)

ax2 = fig.add_subplot(gs[0,1])
ax2.set_xlabel(r"$x$", fontsize=35)
ax2.set_ylabel(r"$z$", fontsize=35)
ax2.scatter(all_x.to(u.kpc), all_z.to(u.kpc), s=.01, c='k', rasterized=True)

ax3 = fig.add_subplot(gs[0,2])
ax3.set_xlabel(r"$y$", fontsize=35)
ax3.set_ylabel(r"$z$", fontsize=35)
ax3.scatter(all_y.to(u.kpc), all_z.to(u.kpc), s=.01, c='k', rasterized=True)

for ax in [ax1,ax2,ax3]:
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)


ax4 = fig.add_subplot(gs[1,:3])
ax4.set_xlabel(r"$\phi_1\ (\degree)$", fontsize=35)
ax4.set_ylabel(r"$\Delta v_r\ (\rm{km\ s^{-1}})$", fontsize=35)
ax4.scatter(all_phi1, straight_vr, s=.1, c='k', rasterized=True)
ax4.scatter(all_phi1[::500], straight_vr[::500], s=30, c='lightcoral')
ax4.set_ylim(-5,5)
ax4.set_xlim(-50,50)


ax5 = fig.add_subplot(gs[:, 3])
ax5.set_xlabel(r"$R$", fontsize=35)
ax5.set_ylabel(r"$z$", fontsize=50)
ax5.plot(R_orbit[:i].to(u.kpc), z_orbit[:i].to(u.kpc), c='k', linewidth=0.5, zorder=0)
sc = ax5.scatter(all_R.to(u.kpc), z_galcen.to(u.kpc), c=all_phi1, s=.01, rasterized=True)
# ax5.set_xlim(10, 28)
# ax5.set_ylim(-20,20)
ax5.set_xlim(2, 19)
ax5.set_ylim(-18,18)
fig.suptitle("%i Myr"%i, fontsize=50)
```
